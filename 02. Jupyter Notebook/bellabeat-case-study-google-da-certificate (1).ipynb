## 3.4 Analyze and Share Phase
### 3.4.1 Definition of correlation
According to Boston University School of Public Health (click the [link](https://sphweb.bumc.bu.edu/otlt/MPH-Modules/PH717-QuantCore/PH717-Module9-Correlation-Regression/PH717-Module9-Correlation-Regression4.html#:~:text=The%20sample%20correlation%20coefficient%20(r,for%20accumulated%20saving%20over%20time.))), the sample correlation coefficient (r) is a measure of the closeness of association of the points in a scatter plot to a linear regression line based on those points.

**Table 2: Description of correlation coefficients**

| Correlation Coefficient (r) | Description (Rough Guideline)  |
| --------------------------- | ------------------------------ |
| +1.0                        | Perfect positive + association |
| +0.8 to 1.0                 | Very strong + association      |
| +0.6 to 0.8                 | Strong + association           |
| +0.4 to 0.6                 | Moderate + association         |
| +0.2 to 0.4                 | Weak + association             |
| 0.0 to +0.2                 | Very weak + or no association  |
| 0.0 to -0.2                 | Very weak - or no association  |
| -0.2 to -0.4                | Weak - association             |
| -0.4 to -0.6                | Moderate - association         |
| -0.6 to -0.8                | Strong - association           |
| -0.8 to -1.0                | Very strong - association      |
| -1.0                        | Perfect negative association   |

### 3.4.2 Analyzing correlations

In this step, we're analyzing correlations between some fields in our tables. 

#### 3.4.2.1 Very active distance vs. Very active minutes

```r
# Analyzing the correlation between Very active distance and very active minutes

minutes_calories_cor <- cor(dailyActivity$veryactiveminutes, dailyActivity$calories)

# Figure 1
ggplot(dailyActivity, aes(x = veryactiveminutes, y = calories)) + 
  geom_point() + 
  geom_smooth(color = "red") +
  labs(title = "Figure 1: Comparison between Very Active Minutes\nand Calories",
       subtitle = paste0("There is a strong correlation (r = ", 
                         round(minutes_calories_cor, digits = 3), ")"))
```

From Figure 1, we can conclude that as long as very active minutes users have, the more calories they burn. In addition, we can see that most users tend to have 50 or less very active minutes. So let's analyze if there is a relationship between these minutes and the very active distances.

#### 3.4.2.2 Very active minutes vs. Very active distance

```r
# Analyzing the correlation between Very active minutes and distance

minutes_distance_cor <- cor(dailyActivity$veryactiveminutes,
                            dailyActivity$veryactivedistance)

# Figure 2
ggplot(dailyActivity, aes(x = veryactivedistance, y = veryactiveminutes)) + 
  geom_jitter() + 
  geom_smooth(color = "purple")+
  labs(title = "Figure 2: Comparison between Very Active Minutes\nand distance",
       subtitle = paste0("There is a very strong correlation (r = ", 
                         round(minutes_distance_cor, digits = 3),")"))
```

As we can see, there is a strong correlation between the very active distance and minutes columns. Also, we can observe that most users have 5 km or less very active distance values. This insight is represented in following histogram.

```r
# Figure 3
ggplot(dailyActivity) + 
  geom_histogram(mapping = aes(x = veryactivedistance, fill = "#f49072"), bins = 10) + 
  scale_x_continuous(breaks = scales::pretty_breaks(n = 10)) +
  labs(title = "Figure 3: Preferences of Very active distance periods") +
  theme(legend.position = "none")
```

From Figure 3, we can conclude that most users tend to have a very active distance from 2 km or less. That insight could be related with working out in gyms (but defining that would require further information) or activities that don't require to transport very far.

#### 3.4.2.3 Calories vs Total steps

Let's analyze if there is a relation between the calories burned and total steps

```r
# Analyzing if there is a correlation between steps and calories burned hourly

calories_steps_cor <- cor(hourlyCaloriesSteps$calories,
                          hourlyCaloriesSteps$steptotal)
# Figure 4
ggplot(hourlyCaloriesSteps, aes(x = steptotal, y = calories)) + 
  geom_jitter() +
  geom_smooth(color = "purple") + 
  labs(title = "Figure 4: Correlation between steps and calories burned hourly",
       subtitle = paste0("There is a very strong correlation (r = ", 
                         round(calories_steps_cor, digits = 3),")"))
```

There is a strong correlation between steps and calories burned. So, as much an user walk, more calories burn.

### 3.4.3 Analyzing times of the day

In this step, I've analyzed the average calories burned and the average amount of steps during the day.

#### 3.4.3.1 Separating our activityhour column into date and time

```r
hourlyCaloriesSteps <- hourlyCaloriesSteps %>% 
  separate(activityhour, into = c("date", "time"), sep = " ") %>% 
  mutate(date = ymd(date))
```

#### 3.4.3.2 Calories and steps during the day

```r
# Calories during the day

# Figure 5
# ggarrange(    I'm not using ggarrange() function because of small size of Kaggle viz                                                                       
hourlyCaloriesSteps %>% group_by(time) %>% 
  summarize(average_calories = mean(calories)) %>% 
  ggplot() +
  geom_col(mapping = aes(x = time, y = average_calories,
                         fill = average_calories)) +
  labs(title = "Figure 5: Average calories burned during the day") +
  scale_fill_gradient(low = "#72f4e9", high = "#f49072") +
  theme(axis.text.x = element_text(angle = 90))
#,

# Figure 6
# Steps during the day

hourlyCaloriesSteps %>% group_by(time) %>% 
  summarize(average_steps = mean(steptotal)) %>% 
  ggplot() +
  geom_col(mapping = aes(x = time, y = average_steps,
                         fill = average_steps)) +
  labs(title = "Figure 6: Average steps made during the day") +
  scale_fill_gradient(low = "#72f4e9", high = "#f49072") +
  theme(axis.text.x = element_text(angle = 90))
#)
```

From Figure 5, we can determine that users burn most calories from 17:00 to 19:00 hours, after that, calories decrease to its minimum at 23:00 until 04:00 hours. In a similar way, according to Figure 6, the hours when more steps are made are from 17:00 to 19:00 hours, after that, number of steps decrease to its minimum at 23:00 until 05:00 hours.

### 3.4.4 Analyzing days of the week

In this step, I've analyzed the average steps, calories burned and hours of sleep by day of the week.

#### 3.4.4.1 Formatting dates as weekdays

```r
#Adding the day of the week

hourlyCaloriesSteps_v2 <- hourlyCaloriesSteps %>% mutate(weekday = weekdays(date)) %>% 
  select(id, date, weekday, calories, steptotal)

# Reordering factor levels by day of the week

hourlyCaloriesSteps_v2$weekday <- ordered(
    hourlyCaloriesSteps_v2$weekday, levels = c(
        "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"))
```

```r
head(hourlyCaloriesSteps_v2)
```

#### 3.4.4.2 Average steps and calories through days of the week

```r
# Calculating average steps and calories burned by day of the week

# Figure 7
#ggarrange(
hourlyCaloriesSteps_v2 %>% group_by(id, date, weekday) %>% 
  summarise(average_steps = sum(steptotal)) %>% group_by(weekday) %>% 
  summarise(average_steps = mean(average_steps)) %>% 
  ggplot() +
  geom_col(mapping = aes(x = weekday, y = average_steps, fill = "#f49072")) +
  labs(title = "Figure 7: Average steps by day of the week") +
  theme(legend.position = "none", axis.text.x = element_text(angle = 45))
#,

# Figure 8
hourlyCaloriesSteps_v2 %>% group_by(id, date, weekday) %>% 
  summarise(average_calories = sum(calories)) %>% group_by(weekday) %>% 
  summarise(average_calories = mean(average_calories)) %>% 
  ggplot() +
  geom_col(mapping = aes(x = weekday, y = average_calories, fill = "#f49072")) +
  labs(title = "Figure 8: Average calories burned by day of the week") +
  theme(legend.position = "none", axis.text.x = element_text(angle = 45))
#)
```

From Figure 7 we can identify that days when users walk more are Tuesday and Saturday. Also, the day with the minimum steps is Sunday. In Figure 8, we can see that calories burned along the week are similar, but they have a descending on Thursday.

#### 3.4.4.3 Determining hours of sleep per day of the week

##### 3.4.4.3.1 Formatting dates as weekdays

```r
# Determining the day of the week when users sleep the best

dailyActivitySleep_v2 <-
dailyActivitySleep %>% mutate(weekday = weekdays(activitydate)) %>%
  select(id, activitydate, weekday, totalminutesasleep, totaltimeinbed)

# Reordering factor levels by day of the week

dailyActivitySleep_v2$weekday <- ordered(
  dailyActivitySleep_v2$weekday, levels = c(
    "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"
  )
)
```

##### 3.4.4.3.2 Plotting my results

```r
# Figure 9

dailyActivitySleep_v2 %>% group_by(weekday) %>% 
  summarise(average_sleep_time = mean(totalminutesasleep)/60) %>% 
  ggplot() +
  geom_col(mapping = aes(x = weekday, y = average_sleep_time, 
                         fill = "#f49072", label = average_sleep_time)) + 
  geom_text(mapping = aes(x = weekday, y = average_sleep_time, 
                          label = round(average_sleep_time,2)), vjust = -0.2) +
    labs(title = "Figure 9: Average time asleep per day") +
  theme(legend.position = "none", axis.text.x = element_text(angle = 45)) +
  xlab("Week of the day") +
  ylab("Average sleep time (in hours)") + 
  ylim(0,8)
```

In Figure 9, we can conclude that the day when users sleep the most is Sunday with an average of 7.55 hours. However, in general, users sleep less than 8 hours, and even, most of them sleep less than 7 hours during the week.
